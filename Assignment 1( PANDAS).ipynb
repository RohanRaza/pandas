# Regions Missing Data Assignment

```python
import pandas as pd
```

```python
wh15=pd.read_csv("wh_2015.csv")
wh16=pd.read_csv("wh_2016.csv")
wh17=pd.read_csv("wh_2017.csv")
regions=pd.read_csv("region.csv")
```

# Filling Regions in 2015

```python
region_2015 = pd.merge(wh15,wh16, on ="Country",how="right",suffixes=("2015","2016"))
```

```python
region_2015.inplace = True
```

```python
region_2015[["Country","Region2015"]].tail(9)
```

# Filling Regions in 2016

```python
regions_2016 = pd.merge(wh15,wh16, on ="Country",how="left",suffixes=("2015","2016"))
```

```python
regions_2016.inplace=True
```

```python
regions_2016[["Country","Region2016"]].tail(9)
```

# Filling Regions in 2017

```python
wh17["Columns"]="Region"
```

```python
region_2017=pd.merge(wh15,wh17,on="Country",how="right",suffixes=("2015","2017"))
```

```python
region_2017[["Country","Region"]].tail(9)
```

```python
regions.rename({"COUNTRY":"Country","REGION":"Region"},axis=1,inplace=True)
```
